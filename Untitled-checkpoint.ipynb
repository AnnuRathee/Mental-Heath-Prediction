```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("mental_health_india_delhi.xlsx")
df
```

```python
print("Dataset Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Entries:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
sns.set(style="whitegrid")
```

```python
# Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Gender", palette="pastel")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("gender_distribution.png")
plt.close()
```

```python
# Mental Health Disorders Count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Mental_Health_Disorder", order=df['Mental_Health_Disorder'].value_counts().index, palette="Set2")
plt.title("Mental Health Disorder Distribution")
plt.xlabel("Disorder Type")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("mental_health_disorder_distribution.png")
plt.close()
```
