```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import pickle
```

```python
df = pd.read_excel("mental_health_india_delhi.xlsx")
df
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.Mental_Health_Disorder.fillna("None",inplace = True)
print(df.isnull().sum())
```

```python
df['Stress_Level'].unique()
```

```python
df['Stress_Level']=df['Stress_Level'].map({'Low':0, 'Moderate':1,  'High':2})
df
```

```python
X = df[['Age', 'Monthly_Income_INR', 'Medical_Expenses_per_Month_INR', 'Sleep_Hours_per_Day', 'Support_System_Score']]
y = df['Stress_Level']
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=77
                                                    )

# Initialize the GaussianNB model
model = GaussianNB()

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
accuracy = model.score(X_test, y_test)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Save the model
with open('gaussian_naive_bayes_model.pkl', 'wb') as f:
    pickle.dump(model, f)
print("Model saved as 'gaussian_naive_bayes_model.pkl'")
```
