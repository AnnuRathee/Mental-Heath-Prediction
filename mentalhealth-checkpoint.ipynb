```python
#  Average Sleep by Stress Level 
avg_sleep = df.groupby('Stress_Level')['Sleep_Hours_per_Day'].mean()
print("\nAverage Sleep Hours per Day by Stress Level:\n", avg_sleep)
```

```python
# Grouped Mean: Income Level vs Support Score 
support_income = df.groupby("Income_Level")["Support_System_Score"].mean()
print("\nAverage Support System Score by Income Level:\n", support_income)
```

```python
# Encode categorical data
from sklearn.preprocessing import StandardScaler, LabelEncoder
label_cols = ['Gender', 'Employment_Status', 'Mental_Health_Disorder', 'Access_to_Counseling', 'Income_Level', 'Stress_Level']
label_encoders = {}
for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# . Hypothesis Testing 
# Compare sleep between high stress (encoded as 2) vs low stress (encoded as 0)
from scipy.stats import ttest_ind
high_stress = df[df['Stress_Level'] == 2]['Sleep_Hours_per_Day']
low_stress = df[df['Stress_Level'] == 0]['Sleep_Hours_per_Day']
t_stat, p_val = ttest_ind(high_stress, low_stress)
print(f"\nT-test (High vs Low Stress Sleep): t={t_stat:.2f}, p={p_val:.4f}")
```

```python
X = df[['Age', 'Monthly_Income_INR', 'Medical_Expenses_per_Month_INR', 'Sleep_Hours_per_Day', 'Support_System_Score']]
y = df['Stress_Level']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=40)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Model Evaluation
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
